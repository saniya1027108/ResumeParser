```python
import PyPDF2
import os
import pandas as pd
```

```python
dir_list = os.listdir(r'/content/drive/MyDrive/writing and translation')
#print(dir_list)

res_list = []
paths = r'/content/drive/MyDrive/writing and translation'
c = 0
for i in dir_list:
    c = c + 1
    if c == 5000:
        break
    pfinal = os.path.join(paths, i)
    pdfReader = PyPDF2.PdfFileReader(pfinal)
    pageObj = pdfReader.getPage(0)

    data = pageObj.extractText()
    res_list.append(data)
```

```python
df=pd.DataFrame(res_list)
df.head(132)
```

```python
df.to_excel('writing.xlsx')
```

```python
df.to_csv('writing.csv')
!cp writing.csv "drive/My Drive/"
```
